## Handling outliers using IQR method

Interquartile Range (IQR) Method
The Interquartile Range (IQR) Method is a statistical technique used to detect and handle outliers in a dataset. The IQR represents the range within which the middle 50% of the data lies, providing a measure of statistical dispersion.

Q1 (First Quartile): The 25th percentile of the data, meaning 25% of the data points are below this value.

Q3 (Third Quartile): The 75th percentile of the data, meaning 75% of the data points are below this value.

IQR=Q3−Q1

```python
q1=df["Passenger_Count"].quantile(0.25)
q1
```

```python
q3=df["Passenger_Count"].quantile(0.75)
q3
```

```python
iqr= q3-q1
iqr
```

```python
lw= q1-0.5*iqr
uw= q3+0.5*iqr
lw,uw
```

```python
df_filterd=df[(df.Passenger_Count>=lw)&(df.Passenger_Count<=uw)]
df_filterd
```

```python
df_filterd.select_dtypes(include="number").skew()
```

skewness reduced.

```python
for column in numeric_columns:
  plt.figure(figsize=(8,4))

  #boxplot
  plt.subplot(1,2,1)
  sns.boxplot(df_filterd[column])
  plt.title(f"{column}-boxplot")
  plt.tight_layout()

  #histogram
  plt.subplot(1,2,2)
  sns.histplot(df_filterd[column],kde=True)
  plt.title(f"{column}-histogram")
  plt.tight_layout()

  plt.show()
```

still there is the presence of outliers.so we are using yeo-johnson method



The Yeo-Johnson method

This is a technique for transforming data to make it more normally distributed. It is a variation of the Box-Cox transformation but can be applied to both positive and negative data, whereas Box-Cox is limited to strictly positive data. The Yeo-Johnson transformation is particularly useful when the data does not follow a normal distribution, as it helps stabilize variance and make the data more suitable for techniques that assume normality, like linear regression or ANOVA.


```python
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method = 'yeo-johnson')
```

# Feature Engineering


Feature engineering is the process of creating new features or transforming existing ones in a dataset to improve the performance of a machine learning model. By enhancing the input features, feature engineering can help the model better capture patterns and relationships, leading to improved predictions.

LABEL ENCODING

Label encoding is a technique used to convert categorical values into numerical values. It is commonly applied in machine learning when the algorithm requires input data to be in a numerical format. Each unique category in a column is assigned an integer label.

Categorical data is label encoded here.

  

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for column in cat_filtered:

  cat_filtered[column] = label_encoder.fit_transform(cat_filtered[column])
print(cat_filtered)




```

Standard scaler

Standard scaler is a technique used in data preprocessing to scale features so that they have a mean of 0 and a standard deviation of 1. It is one of the most common types of feature scaling used in machine learning, ensuring that each feature contributes equally to the model.

we scale numerical data here

SelectKBest is a feature selection method used in machine learning to select the top K features from a dataset based on their relevance or statistical significance to the target variable. # Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,root_mean_squared_error
from sklearn.model_selection import GridSearchCV
```

Regression models are used to predict a continuous outcome variable (also called a dependent variable) based on one or more predictor variables (also called independent variables). There are several types of regression models, depending on the nature of the data and the relationships between variables. following are the different models.



1.   Linear regression is one of the simplest and most commonly used models for predicting a continuous outcome. It assumes a linear relationship between the dependent variable (what you're trying to predict) and one or more independent variables (the predictors). The goal of linear regression is to find the best-fitting straight line (regression line) through the data points.
2.   Decision Tree Regression
A tree-like model where decisions are made based on certain conditions, with continuous values predicted at the leaf nodes.

3.  Random Forest Regression
An ensemble method that builds multiple decision trees and averages their predictions to improve accuracy.
4.   Gradient Boosting Regression
These are advanced tree-based algorithms that iteratively improve by minimizing the prediction error through gradient boosting.


5.   Support Vector Regression (SVR)
A regression model that uses Support Vector Machines to find a hyperplane that best fits the data points.








```python
models={
    '1.Linear Regression':LinearRegression(),
    '2.Decision Tree Regressor':DecisionTreeRegressor(),
    '3.Random Forest Regressor':RandomForestRegressor(),
    '4.Gradient Boosting Regressor':GradientBoostingRegressor(),
    '5.Support Vector Regressor':SVR()
}
```

```python
result={}
for model_name,model in models.items():
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    mae=mean_absolute_error(y_test,y_pred)
    mse=mean_squared_error(y_test,y_pred)
    rmse=np.sqrt(mse)
    r2=r2_score(y_test,y_pred)
    result[model_name]={'mae':mae,'mse':mse,'rmse':rmse,'r2':r2}
```

```python
result_df=pd.DataFrame(result).T
pd.options.display.float_format='{:.4f}'.format
result_df
```

observation


**Random Forest Regressor:**


mae-0.0936,mse-0.03556,rmse-0.1887,r2-0.9639

The MAE and RMSE are the lowest among all models, indicating that Random Forest is making the smallest errors in prediction.
The R² value of 0.96 shows that it explains 96.62% of the variance in the data, making it the best-performing model overall. Random Forest, being an ensemble of Decision Trees, performs better due to averaging, reducing variance, and preventing overfitting.

```python
model = RandomForestRegressor()

model.fit(x_train, y_train)

y_train_pred = model.predict(x_train)
y_test_pred = model.predict(x_test)

train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)

train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

print(f"Training MSE: {train_mse}, Testing MSE: {test_mse}")
print(f"Training R²: {train_r2}, Testing R²: {test_r2}")
```

# Hyperparameter tuning

Hyperparameter tuning is the process of optimizing the hyperparameters of a machine learning model to improve its performance. Unlike model parameters (like coefficients in linear regression), hyperparameters are not learned from the data directly but must be set before training the model. Examples of hyperparameters include the number of trees in a Random Forest, learning rate in Gradient Boosting, and the penalty parameter in Support Vector Machines.

Grid search

Description: Grid search is an exhaustive search technique where you specify a grid of hyperparameter values and test all possible combinations.

```python
best_model = RandomForestRegressor(n_estimators=50,random_state=42,max_depth=20, min_samples_split=5, min_samples_leaf=2, max_features='sqrt',
                                   n_jobs=-1   )
best_model.fit(x_train,y_train)
```

```python

param_grid = {
   'n_estimators': [50, 100, 200,300],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]}
```

```python
grid_search = GridSearchCV(
   estimator=RandomForestRegressor(),
    param_grid=param_grid,
   cv=10,
    scoring='r2',
    n_jobs=-1,
    verbose=2
)


```

```python
grid_search.fit(x_train,y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best R² Score:", grid_search.best_score_)
```

```python
best_model=grid_search.best_estimator_
best_model
```
